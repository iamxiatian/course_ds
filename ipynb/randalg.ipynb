
# 随机化算法（Randomized Algorithm）

## 1 引言

随机化算法在现代计算机科学领域占据着举足轻重的地位。相较于传统确定性算法，随机化算法凭借独特的设计理念，为诸多复杂问题的解决开辟了新路径。

### 随机化算法的优势​


随机化算法是相对于确定性算法而言的。对于确定性算法（Deterministic Algorithm），
输出以及运行时间均仅为输入的函数 ，也就是说对于给定输入，确定性算法每次运行输出
和耗时是确定、仅由输入决定的。 比如在基础排序算法里，输入一组数，确定性排序算法
（像冒泡排序）的输出（排好序的序列）和运行时间（基于输入数据规模和初始顺序等，
是确定由输入决定的计算耗时情况 ）是相对固定、仅由输入本身特性决定的 ，和随机因素
无关 。 

![确定性算法](figures/deterministic.png)

随机化算法在给定搜索空间内，随机选择一个元素，并尝试扩展该元素，直到找到一个解
或者搜索空间为空，因此算法的输出和运行时间会随输入有一定的随机性。

随机化算法的优势主要体现在两个关键方面：

其一，性能卓越。在众多实际问题中，随机化算法能够以超越已知最优确定性算法的速度完成任务。以排序问题为例，传统的确定性排序算法，如冒泡排序，在最坏情况下的时间复杂度高达 $O(n^2)$。而随机化快速排序算法，通过随机选择基准元素，期望时间复杂度稳定在$O(n \log n)$，在平均情况下性能远超冒泡排序。​

其二，实现简洁。随机化算法的逻辑架构往往更为简单直观，这使得其在描述和实现过程中更易于理解与操作。在处理一些复杂的数据结构操作时，随机化算法能够以较少的代码量和更清晰的逻辑实现相同甚至更好的功能，降低了程序开发与维护的难度。​

### 对抗者模型与随机化算法​
对抗者模型是理解随机化算法的核心概念。在确定性算法的情境下，对抗者能够精心构造特定输入，使得算法性能急剧恶化。例如，对于快速排序算法，如果输入数据恰好是完全有序或者逆序的，快速排序的时间复杂度将退化为 $O(n^2)$，这便是对抗者通过特定输入对确定性算法造成的负面影响。​

与之不同，随机化算法可被视作在一组确定性算法上构建的概率分布。由于算法执行过程中引入了随机因素，对抗者难以通过构造单一输入来有效阻碍算法运行。以随机化快速排序来说，每次选择基准元素的随机性，使得对抗者无法像针对普通快速排序那样，通过特定输入让算法陷入最坏情况，从而显著提升了算法在各种输入场景下的稳定性与平均性能。

## 2 排序算法中的随机化应用

排序问题的目标是将含 n 个元素的数组按升序排列（输入为数组$S$，输出为满足 $S[i] \leq S[i+1]（i=1,...,n-1）$的排列）。

### 快速排序：高效但脆弱的确定性排序

快速排序基于分治思想，是一种高效的排序算法，并且具有原地排序的特性，即不需要额外的大量辅助空间。其主要步骤如下：​

选基准：通常选取数组的首元素作为基准元素。这个基准元素将作为划分数组的关键依据，其选择的合理性对算法性能有着重要影响。​

分区：以选定的基准元素为标准，将数组划分为两部分：$S_1$，其中的元素均小于或等于基准元素；$S_2$，其中的元素均大于基准元素。在分区过程中，通过指针移动和元素交换操作，在 $O(n)$的时间复杂度内完成数组的分区，使得数组元素依据与基准元素的大小关系被大致分成两个有序的子数组。​

递归排序：对$S_1$和$S_2$两个子数组分别递归执行快速排序。通过不断递归，逐渐将各个子数组排序，最终使得整个数组达到有序状态。​

在理想情况下，若每次分区都能将数组均匀地分为两部分，快速排序的递归关系为$T(n) = 2T (n/2) + n$，时间复杂度为 $O (n \log n)$，展现出极高的效率。然而，快速排序的性能对输入数据的分布极为敏感。当对抗者构造出逆序数组作为输入时，每次分区都会出现极端不平衡的情况，即一边只有 1 个元素，另一边有 n - 1 个元素。此时，快速排序的递归关系变为 $T(n) = T(n - 1) + n$，时间复杂度退化为$O(n^2)$，性能急剧下降，这凸显了快速排序在面对特定输入时的脆弱性。

### 随机化快速排序：引入随机因素提升稳定性

随机化快速排序在普通快速排序的基础上，引入了关键的随机化机制，即随机选择基准元素。其核心步骤如下：​

随机选基准：从数组$S$中均匀随机地选取一个元素$Y$作为基准元素。这种随机选择方式打破了普通快速排序中基准元素选择的确定性，使得算法在面对各种输入数据时具有更强的适应性。​

分区：与普通快速排序类似，以随机选定的基准元素$Y$为界，将数组$S$分为$S_1$（元素小于$Y$）和$S_2$（元素大于$Y$）两部分。通过合理的指针操作和元素交换，在$O(n)$时间内完成分区任务。​

递归排序：对$S_1$和$S_2$两个子数组递归进行排序。最后，将$S_1$的排序结果、基准元素$Y$和$S_2$的排序结果依次拼接起来，得到整个数组的有序排列。​

在期望时间复杂度分析方面，以比较次数作为衡量算法成本的主要指标。定义随机变量 Xᵢⱼ，当第 i 小元素 S (i) 与第 j 小元素 S (j) 进行比较时，Xᵢⱼ取值为 1，否则为 0。经过深入分析可知，S (i) 与 S (j) 进行比较的概率为 2/(j - i + 1)。这是因为只有当两者之一在其 “中间元素” 前被选为基准时，它们才会进行比较。通过对所有可能的元素对进行求和，即总期望比较次数为 ΣΣ2/(j - i + 1)。利用调和数 Hₙ≈ln n 的性质进行化简，最终可以得出随机化快速排序的期望时间复杂度为 O (n log n)。这一结果具有重要意义，它表明随机化快速排序的期望性能仅依赖于算法自身的随机选择机制，而与输入数据的分布情况无关。无论输入数据呈现何种分布特征，随机化快速排序都能以较高的概率保持高效的运行效率，极大地提升了算法的稳定性和实用性。


## 拉斯维加斯与蒙特卡洛算法：随机化算法的两大类别​
### 拉斯维加斯算法的特性与示例​
拉斯维加斯算法是随机化算法中的一类重要算法，其核心特性在于总能给出正确的结果，然而，算法的运行时间是一个随机变量。以随机化快速排序算法为例，它在期望情况下的时间复杂度为 O (n log n)，这意味着在大多数情况下，算法能够以高效的时间完成排序任务。但在最坏情况下，其时间复杂度可能退化为 O (n²)，不过这种最坏情况出现的概率极低。​

若一个拉斯维加斯算法的期望运行时间能够控制在输入规模的多项式函数范围内，那么我们称该算法是高效的。在实际应用中，拉斯维加斯算法常用于对结果正确性要求极高，且能够接受一定时间波动的场景。例如，在一些金融交易风险评估模型中，需要确保计算结果的绝对准确，同时对于计算时间在合理范围内的波动具有一定容忍度，此时拉斯维加斯算法就能够发挥其独特优势。​

### 蒙特卡洛算法的特性与示例​
蒙特卡洛算法与拉斯维加斯算法有所不同，它可能会输出错误的结果，但通过巧妙的设计，其错误概率可以通过重复执行算法来有效降低。以最小割算法为例，每次执行该算法得到的候选最小割可能并非真正的最小割，但随着执行次数的增加，找到真正最小割的概率会不断提高。​

若蒙特卡洛算法的最坏运行时间能够控制在输入规模的多项式函数范围内，那么该算法也被认为是高效的。在实际应用中，蒙特卡洛算法常用于那些对计算效率要求较高，且能够容忍一定错误率的场景。例如，在一些大规模数据的模拟分析场景中，由于数据量巨大，精确计算往往需要耗费大量时间和资源，此时蒙特卡洛算法可以在较短时间内给出一个具有一定可信度的近似结果，为决策提供重要参考。​

### 两类算法的比较与选择​
拉斯维加斯算法和蒙特卡洛算法在特性上存在明显差异。从结果正确性角度看，拉斯维加斯算法总能保证结果正确，而蒙特卡洛算法存在一定错误概率；从运行时间角度看，拉斯维加斯算法的运行时间是随机变量，蒙特卡洛算法的运行时间在最坏情况下具有确定性。实际上，拉斯维加斯算法可以看作是错误概率为 0 的蒙特卡洛算法。​
在实际应用中，选择使用哪种算法需要根据具体问题的需求来决定。如果问题对结果的正确性要求极高，不容许出现任何错误，那么拉斯维加斯算法是首选；如果问题对计算效率要求较高，且能够接受一定程度的错误率以换取更快的计算速度，那么蒙特卡洛算法更为合适。例如，在密码学领域，由于对安全性和结果准确性的严格要求，通常会优先选择拉斯维加斯算法；而在一些数据挖掘和机器学习的预处理阶段，为了快速对大规模数据进行初步分析和筛选，蒙特卡洛算法因其高效性而更受青睐。
